```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load the cleaned data from Phase 1
df = pd.read_csv('data/cleaned/cleaned_data.csv')

# Convert text categories into 1s and 0s (One-Hot Encoding) so the computer can read them
df_phase2 = pd.get_dummies(df, columns=['Street', 'Central Air'])

# Convert Quality ratings (Ex, Gd, TA) into a 1-5 scale
qual_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1}
df_phase2['Exter Qual Num'] = df['Exter Qual'].map(qual_map).fillna(3)

# Group houses by age categories to see if "New" vs "Old" affects price
df_phase2['Age_Group'] = pd.cut(df['Year Built'], bins=[0, 1950, 1990, 2010], labels=['Old', 'Mid', 'New'])

# Create a 'Price Per Room' feature to measure housing value efficiency
df_phase2['Price_Per_Room'] = df['SalePrice'] / (df['TotRms AbvGrd'] + 1)

# Combine Size and Quality into one "Super Feature" to see the total impact on value
df_phase2['Size_Quality'] = df['Gr Liv Area'] * df['Overall Qual'].astype(int)

# Scale the Area columns so that large numbers don't overwhelm the smaller ones in the math
scaler = StandardScaler()
df_phase2[['Lot Area', 'Gr Liv Area']] = scaler.fit_transform(df_phase2[['Lot Area', 'Gr Liv Area']])

# Save the final engineered dataset for analysis
df_phase2.to_csv('data/cleaned/featured_data.csv', index=False)
print("Phase 2 complete: Features created and scaled!")
```
