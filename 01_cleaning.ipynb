```python
import pandas as pd
import numpy as np

# Load the raw dataset from the project folder
df = pd.read_csv('data/raw/AmesHousing.csv')

def clean_data(raw_df):
    # Prevent warning messages when handling data types
    pd.set_option('future.no_silent_downcasting', True)
    df_clean = raw_df.copy()
    
    # Convert ID-like numbers to 'object' (categories) because we won't do math on them
    df_clean['MS SubClass'] = df_clean['MS SubClass'].astype('object')
    df_clean['Overall Qual'] = df_clean['Overall Qual'].astype('object')
    
    # Remove any identical rows to keep data unique
    df_clean = df_clean.drop_duplicates()
    
    # Fill missing numbers with the Median (middle value) to avoid skewing results
    num_cols = df_clean.select_dtypes(include=[np.number]).columns
    for col in num_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())
    
    # Fill missing text/categories with the Mode (most frequent value)
    cat_cols = df_clean.select_dtypes(include=['object']).columns
    for col in cat_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
    
    # Handle Outliers: Cap the top 1% of house prices to the 99th percentile 
    # This prevents extreme mansions from ruining the average
    target = 'SalePrice'
    q99 = df_clean[target].quantile(0.99)
    df_clean[target] = np.where(df_clean[target] > q99, q99, df_clean[target])
    
    return df_clean

# Run the function and save the result for Phase 2
df_phase1 = clean_data(df)
df_phase1.to_csv('data/cleaned/cleaned_data.csv', index=False)
print("Phase 1 complete: Data is clean and saved!")
```
