```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the final engineered dataset created in Phase 2
df = pd.read_csv('data/cleaned/featured_data.csv')

# --- VISUALIZATION 1: Histogram ---
# Create a figure for the first chart
plt.figure(figsize=(8, 5))
# Plot a histogram to see the distribution of house prices
# 'kde=True' adds a smooth line (Kernel Density Estimate) to show the shape of the data
sns.histplot(df['SalePrice'], color='blue', kde=True)
plt.title('Distribution of House Prices in Ames')
plt.show()

# --- VISUALIZATION 2: Scatter Plot ---
# Create a figure for the second chart
plt.figure(figsize=(8, 5))
# Plot Living Area vs. Sale Price
# 'hue' colors the dots by Overall Quality to see how quality affects price
sns.scatterplot(x='Gr Liv Area', y='SalePrice', hue='Overall Qual', data=df)
plt.title('Living Area (Scaled) vs. Sale Price')
plt.show()

# --- MANUAL STATISTICAL CALCULATIONS ---
# Convert the SalePrice column into a NumPy array for math operations
prices = df['SalePrice'].values

# Calculate the Mean (Average) manually: Sum of all prices divided by the count
my_mean = np.sum(prices) / len(prices)

# Calculate the Standard Deviation (Spread) manually:
# 1. Find the square of the difference between each price and the mean
# 2. Find the average of those squared differences (Variance)
# 3. Take the square root to get the Standard Deviation
my_std = np.sqrt(np.sum((prices - my_mean)**2) / len(prices))

# Output the final results for the report
print(f"Average House Price: ${my_mean:,.2f}")
print(f"Price Spread (Std Dev): ${my_std:,.2f}")
```
